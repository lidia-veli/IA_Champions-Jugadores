```python
import pandas as pd
```

## Partidos 2023

```python
df_23 = pd.read_csv('resultados_partidos_2023.csv')
df_23.head()
```

```python
# eliminar columnas Team_ID
df_23 = df_23.drop(['Team_ID_Eq1', 'Team_ID_Eq2'], axis=1)

# aÃ±adir columna Match_ID
match_id = pd.Series(range(1, df_23.shape[0]+1))
df_23['Match_ID'] = match_id.astype(str)
df_23.set_index('Match_ID', inplace=True)

df_23.head(3)
```

## Partidos 2024

```python
df_24 = pd.read_csv('resultados_partidos_2024.csv')
df_24.head()
```

```python
# eliminar columnas Team_ID
col_id = ['Team_ID_Eq1', 'Team_ID_Eq2']
for col in df_24.columns.to_list():
    if col in col_id:
        df_24.drop(col, axis=1, inplace=True)
    pass
```

```python
df_24.dtypes.value_counts()
```

```python
len_23 = df_23.shape[0]
print(len_23)
len_24 = df_24.shape[0]
print(len_24)
```

```python
# aÃ±adir columna Match_ID
match_id = pd.Series(range(len_23+1, len_23+len_24+1))
df_24['Match_ID'] = match_id.astype(str)
df_24.set_index('Match_ID', inplace=True)

df_24.head(3)
```

## Juntar 23 y 24

```python
if df_23.columns.equals(df_24.columns):
    print('Los dataframes tienen las mismas columnas')
    df_23.to_csv('partidos_2023.csv', index=True)
    df_24.to_csv('partidos_2024.csv', index=True)
else:
    print('Los dataframes no tienen las mismas columnas')
```

```python
df_23_24 = pd.concat([df_23, df_24], )
df_23_24.to_csv('todos_partidos.csv', index=True)
```
