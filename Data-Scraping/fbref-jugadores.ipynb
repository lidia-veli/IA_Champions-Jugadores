```python
dict_stats = scraper.scrape_all_stats(year=2024, league='Champions League', normalize=False)
# KEYS: nombre categorias
# VALUES: tupla de dataframes (squad_stats, opponent_stats, player_stats)
```

```python
stat_cat = ['standard', 'goalkeeping', 'advanced goalkeeping', 'shooting', 
            'passing', 'pass types', 'goal and shot creation', 'defensive', 'possession', 'playing time', 'misc']

categ_estadisticas = ['standard', 'goalkeeping', 'shooting', 'goal and shot creation']
```

```python
dict_stats['standard'][0]
```

```python
def reasignar_nombres_columnas(df):
    # obtener nombres columnas
    nombres_col = []
    for tup in df.columns[:-2]:
        n1, n2 = tup
        nombres_col.append(n2)

    for tup in df.columns[-2:]: # player link y player ID
        n1, n2 = tup
        nombres_col.append(n1)

    print( nombres_col )
    print(len(nombres_col))

    # reasignar nombres en el dataframe
    df.columns = nombres_col

    return df



dict_df_eq = {}  # dict por categorias de los equipos
dict_df_jug = {}

def sacar_df_categ(categorias, dict_stats):

    for cat in categorias:
        # EQUIPOS -------------------
        # sacar df del diccionario
        df_eq = dict_stats[cat][0]
        # renombrar columnas
        df_eq = reasignar_nombres_columnas(df_eq)
        # asignar df al diccionario por categoria
        dict_df_eq[cat] = df_eq

        # JUGADORES -----------------
        # sacar df del diccionario
        df_jug = dict_stats[cat][2]
        # renombrar columnas
        df_jug = reasignar_nombres_columnas(df_jug)
        # asignar df al diccionario por categoria
        dict_df_jug[cat] = df_jug
```

```python
sacar_df_categ(categ_estadisticas, dict_stats)
```
